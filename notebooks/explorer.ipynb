# IPO & ETF Launch Tracker — Data Explorer

Connect to the SQLite database and interactively explore all pipeline tables.  
Run cells top-to-bottom, or jump to any section you need.

```python
import sqlite3
import pandas as pd
from pathlib import Path

# Display settings
pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", 60)
pd.set_option("display.width", 200)

# Connect to database
DB_PATH = Path("../db/ipo_etf_tracker.db")
assert DB_PATH.exists(), f"Database not found at {DB_PATH.resolve()}. Run 'python main.py --init' first."
conn = sqlite3.connect(str(DB_PATH))
print(f"Connected to {DB_PATH.resolve()}")

tables = [r[0] for r in conn.execute("SELECT name FROM sqlite_master WHERE type='table'").fetchall()]
print(f"Tables: {tables}")
```

## Stock IPOs Pipeline

All entities from S-1/F-1 filings, tracking from registration through launch.

```python
# All stock IPOs
df_ipos = pd.read_sql_query("SELECT * FROM stock_ipos ORDER BY updated_at DESC", conn)
print(f"Total IPO entities: {len(df_ipos)}")
df_ipos
```

```python
# Filter IPOs by status (change the status value as needed)
status_filter = "FILED"  # Options: FILED, AMENDED, PRICED, EFFECTIVE, LAUNCHED
df_ipos_filtered = pd.read_sql_query(
    "SELECT * FROM stock_ipos WHERE status = ? ORDER BY filing_date DESC",
    conn, params=[status_filter]
)
print(f"IPOs with status '{status_filter}': {len(df_ipos_filtered)}")
df_ipos_filtered
```

## ETF Launches Pipeline

All entities from N-1A and related filings, tracking from registration through launch.

```python
# All ETF launches
df_etfs = pd.read_sql_query("SELECT * FROM etf_launches ORDER BY updated_at DESC", conn)
print(f"Total ETF entities: {len(df_etfs)}")
df_etfs
```

```python
# Filter ETFs by status
status_filter = "FILED"  # Options: FILED, AMENDED, PRICED, EFFECTIVE, LAUNCHED
df_etfs_filtered = pd.read_sql_query(
    "SELECT * FROM etf_launches WHERE status = ? ORDER BY filing_date DESC",
    conn, params=[status_filter]
)
print(f"ETFs with status '{status_filter}': {len(df_etfs_filtered)}")
df_etfs_filtered
```

## Filing Events Log (Audit Trail)

Every individual SEC filing is recorded here. This is the raw event history — one row per filing document.

```python
# Filing events with entity names
df_events = pd.read_sql_query("""
    SELECT
        fe.id,
        fe.entity_type,
        fe.entity_id,
        CASE
            WHEN fe.entity_type = 'IPO' THEN si.entity_name
            WHEN fe.entity_type = 'ETF' THEN el.fund_name
        END AS entity_name,
        fe.form_type,
        fe.filing_date,
        fe.accession_number,
        fe.description,
        fe.created_at
    FROM filing_events fe
    LEFT JOIN stock_ipos si ON fe.entity_type = 'IPO' AND fe.entity_id = si.id
    LEFT JOIN etf_launches el ON fe.entity_type = 'ETF' AND fe.entity_id = el.id
    ORDER BY fe.filing_date DESC
""", conn)
print(f"Total filing events: {len(df_events)}")
df_events
```

## Active Securities (Launched)

Entities that have been declared effective and moved to the action table.

```python
# Active securities
df_active = pd.read_sql_query("SELECT * FROM active_securities ORDER BY created_at DESC", conn)
print(f"Total active securities: {len(df_active)}")
df_active
```

## Notifications

Launch alerts. Unread notifications indicate newly launched securities you haven't reviewed yet.

```python
# Unread notifications
df_notif = pd.read_sql_query(
    "SELECT * FROM notifications WHERE is_read = 0 ORDER BY created_at DESC", conn
)
print(f"Unread notifications: {len(df_notif)}")
df_notif
```

## Summary Statistics

Status distribution across all IPO and ETF entities.

```python
# Status counts for IPOs and ETFs
print("=== Stock IPO Status Distribution ===")
df_ipo_status = pd.read_sql_query(
    "SELECT status, COUNT(*) as count FROM stock_ipos GROUP BY status ORDER BY count DESC", conn
)
print(df_ipo_status.to_string(index=False))

print("\n=== ETF Launch Status Distribution ===")
df_etf_status = pd.read_sql_query(
    "SELECT status, COUNT(*) as count FROM etf_launches GROUP BY status ORDER BY count DESC", conn
)
print(df_etf_status.to_string(index=False))

print(f"\n=== Totals ===")
print(f"IPO entities: {len(df_ipos) if 'df_ipos' in dir() else '(run IPO cell first)'}")
print(f"ETF entities: {len(df_etfs) if 'df_etfs' in dir() else '(run ETF cell first)'}")
print(f"Filing events: {conn.execute('SELECT COUNT(*) FROM filing_events').fetchone()[0]}")
print(f"Active securities: {conn.execute('SELECT COUNT(*) FROM active_securities').fetchone()[0]}")
print(f"Unread notifications: {conn.execute('SELECT COUNT(*) FROM notifications WHERE is_read=0').fetchone()[0]}")
```
